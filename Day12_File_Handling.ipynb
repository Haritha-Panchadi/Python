### Files in python

- Open() ->
  - r ->Read the file from existing file
  - w ->Write data to new file/if file exist remove all the data and new data
  - a -> write the data
- Do some Operations
  - read the data
  - write the data
- Close()  

```python
f = open('file.txt','w')

f.close()
```

```python
for i in range(1,11):
  f = open('file' + str(i) + '.txt', 'w')

  f.close()
```

```python
f= open('file.txt', 'w')
for i in range(1,101):
  f.write('21APSSDC' + str(i) + '\n')
f.close()
```

```python
f= open('file.txt', 'a')
for i in range(1,101):
  f.write('22APSSDC' + str(i) +'******')
f.close()
```

```python
f= open('file.txt', 'r')

data = f.readlines()

for line in data:
  print(line)

f.close()
```

- r+ (read as well as append the file)
- w+ (write as well as read the file)
- a+ (append as well as read the file)

```python
f= open('file.txt', 'r+')

data = f.read()
f.write(data)

```

```python
f= open('file.txt', 'r+')

data = f.read()
f.write(data)
f.close()
```
